```python
import numpy as np
import pandas as pd
from scipy import linalg
import scipy
a = [np.random.randint(1,100,50) for i in range(5)]
```

```python
a = np.matrix(a)
a.shape
```

```python

a = np.matrix(a)
K = 4
#mean normalization
am = np.zeros(shape=a.shape)
for i in range(a.shape[1]):
    for j in range(a.shape[0]):
        am[j,i] = a[j,i] - np.mean(a[:,i])
    
# from sklearn.preprocessing import scale
# X = scale( X, axis=0, with_mean=True, with_std=True, copy=True )

cov = np.dot(np.transpose(am),am)*(1/am.shape[0])
U, s, Vh = linalg.svd(cov)
```

```python
z= np.zeros(shape=(am.shape[0],K))
U = np.matrix(U)

for j in range(K):
    for i in range(am.shape[0]):
        z[i,j] = np.transpose(am[i,:]) * U[:,j]
      
print(z)

z = np.matrix(z)
U = np.matrix(U)
U
```

```python
#recover
ax = np.zeros(shape=a.shape)
for i in range(z.shape[0]):
    for j in range(U.shape[0]):
        ax[i,j] = z[i,:] * np.transpose(U[j,0:K])
print(ax)
print(am)
```

```python
#average squared projection error
sqp = (1/am.shape[0]) * sum(np.linalg.norm(am - ax ,axis=0)**2)    
sqp
```

```python
#Total variation in the data
TV = (1/am.shape[0]) * sum(np.linalg.norm(am,axis=0)**2)    
TV
```

```python
#ratio of sqp and TV should be less than or equal to 0.01 to keep the 99 percent variance.
sqp/TV
```

```python
# to check the variance in a simple way use the sum of s value from the SVD 
#and take first k number of s devide it by all sum of s it should be greater than or equal to 0.99
sum(s[0:K])/sum(s)
```

```python
# PCA using sikitlearn.
from sklearn.decomposition import PCA
pca = PCA(n_components=K)
pca.fit(a)

print(pca.transform(a))
print(pca.explained_variance_ratio_) 
print(pca.explained_variance_)
print(pca.singular_values_)
print(pca.components_)
pca_score = pca.explained_variance_ratio_
V = pca.components_
```
